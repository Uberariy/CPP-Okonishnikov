**Задача 2.**  Норма обязательных резервов равна 20%. Коммерческий банк хранит еще 5% от суммы 
депозитов в качестве избыточных резервов. Величина депозитов составляет 10 000. Какую 
максимальную сумму банк может использовать для выдачи ссуд?

```python
# Величина обязательных резервов:
A = 10000 * (20/100)

# Избыточные резервы:
B = 10000 * (5/100)

# Суммарные резервы:
R = A + B

print("Ответ:", 10000 - R)
```

используют свои кредитные возможности, при условии, что норма резервирования депозитов
