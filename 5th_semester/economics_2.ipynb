**Задача 4.** Длина пляжа 700 метров. Продавец А расположил свой магазин на расстоянии 
150 метров от левого конца пляжа, а продавец В расположил магазин на расстоянии 350 
метров от правого конца пляжа. Что можно сказать о соотношении цен в магазинах А и В 
на одинаковую продукцию?

Предположим, что отдыхающие на пляже распределены равномерно и посещают ближайший к месту своего отдыха ларёк. Понятно, что 150 метров от левого конца пляжа будут "покрываться" продавцом А, а 350 метров от правого конца продавцом B. Оставшая цетральная зона поделиться между двумя продавцами по 100 метров на каждого. Получается, что в теории второй продавец оказывается в более выгодном положении и волен установить цену **в 1.8 раз больше**, чем первый на одинаковую продукцию.

**Задача 5.** Издержки конкурентной фирмы равны: TC = 20 + 33q – 4q^2 + q^3
. При каких ценах 
продукта целесообразно продолжать производство?


TVC = TC - TFC = 33q - 4q^2 + q^3  
AVC = TVC/q = 33 - 4q + q^2

Прибыль максимальна при P == MC;   
MC = dTVC/dq = 33 - 8q + 3q^2

Целосообразно продолжать производство при P>AVC
  => При P = ABC уже нецелесообразно. Получим уравнение:

AVC = MC   
33 - 4q + q^2 = 33 - 8q + 3q^2   
=> q = 2    
=> P = 33 - 16 + 12 = 29

**Ответ: 29**


**Задача 6.** Условия рыночного спроса выражены уравнением Q=30-2P. Для каждой из двух 
фирм олигополистов функция общих издержек имеет вид TC=3Q. Если фирмы 
взаимодействуют по Курно, то выпуск каждой из них составит?


P = (30 - Q)/2 == a - bQ   
=> a = 15    
=> b = 0.5

TC = 3Q == cQ    
=> c = 3

Выпуск каждой из них:    
Qboth = 2(15-3)/(3*0.5) = 16   
Qeach = 16/2 = 8

**Ответ: 8**
